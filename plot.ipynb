```python
import pandas as pd
#https://meteostat.net/en/place/de/dresden?s=D1051&t=2022-01-01/2022-12-31
df = pd.read_csv('dresden.csv')
df['date'] = pd.to_datetime(df.date)
# df.set_index(['date'], inplace=True)
# print(df)
```

```python
df.plot(y='prcp')
```

```python
wk = df.groupby(df['date'].dt.day_name()).mean()
plt = wk.plot.bar(y='prcp', title='Average rain per week day [mm]')
plt.figure.savefig('average.svg')
df['rain'] = df['prcp'] > 0.05
wksum = df.groupby(df['date'].dt.day_name()).sum()
plt2 = wksum.plot.bar(y='prcp', title='Days with rain in 2022')
plt2.figure.savefig('days.svg')
```
